```python
# Dependencies and Setup
import pandas as pd
from pathlib import Path

# File to Load (Remember to Change These)
school_data_to_load = Path("Resources/schools_complete.csv")
student_data_to_load = Path("Resources/students_complete.csv")

# Read School and Student Data File and store into Pandas DataFrames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset.  
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
school_data_complete.head()
```

## District Summary

```python
# Calculate the total number of unique schools
unique_school_count = school_data_complete['school_name'].unique()
school_count = len(unique_school_count)
school_count
```

```python
# Calculate the total number of students
student_count = school_data_complete["student_name"].count()
student_count
```

```python
# Calculate the total budget
total_budget = school_data["budget"].sum()
total_budget
```

```python
# Calculate the average (mean) math score
average_math_score = school_data_complete["math_score"].mean()
average_math_score
```

```python
# Calculate the average (mean) reading score
average_reading_score = school_data_complete["reading_score"].mean()
average_reading_score
```

```python
# Use the following to calculate the percentage of students who passed math (math scores greather than or equal to 70)
passing_math_count = school_data_complete.loc[school_data_complete["math_score"] >= 70]
students_passing_math_percent = passing_math_count["Student ID"].count()
passing_math_percentage = (students_passing_math_percent / student_count) * 100
passing_math_percentage
```

```python
# Calculate the percentage of students who passed reading (hint: look at how the math percentage was calculated)  
passing_reading_count = school_data_complete.loc[school_data_complete["reading_score"] >= 70]
students_passing_reading_percent = passing_reading_count["Student ID"].count()
passing_reading_percentage = (students_passing_reading_percent / student_count) * 100
passing_reading_percentage
```

```python
# Use the following to calculate the percentage of students that passed math and reading
passing_math_reading_count = school_data_complete[(school_data_complete["math_score"] >= 70) & (school_data_complete["reading_score"] >= 70)]
students_passing_rate = passing_math_reading_count["reading_score"].count()
overall_passing_rate = students_passing_rate / student_count * 100
overall_passing_rate
```

```python
# Create a high-level snapshot of the district's key metrics in a DataFrame
district_summary = {
    "Total Schools": [school_count],
    "Total Students": [student_count],
    "Total Budget": [total_budget],
    "Average Math Score": [average_math_score],
    "Average Reading Score": [average_reading_score],
    "% Passing Math": [passing_math_percentage],
    "% Passing Reading": [passing_reading_percentage],
    "% Overall Pass": [overall_passing_rate]}

district_summary_df = pd.DataFrame(district_summary)

# Formatting
district_summary_df["Total Students"] = district_summary_df["Total Students"].map("{:,}".format)
district_summary_df["Total Budget"] = district_summary_df["Total Budget"].map("${:,.2f}".format)
district_summary_df["Average Math Score"] = district_summary_df["Average Math Score"].map("{:.4f}".format)
district_summary_df["Average Reading Score"] = district_summary_df["Average Reading Score"].map("{:.4f}".format)
district_summary_df["% Passing Math"] = district_summary_df["% Passing Math"].map("{:.4f}".format)
district_summary_df["% Passing Reading"] = district_summary_df["% Passing Reading"].map("{:.4f}".format)
district_summary_df["% Overall Pass"] = district_summary_df["% Overall Pass"].map("{:.4f}".format)

# Display the DataFrame
district_summary_df
```

```python
# Calculate the total student count per school
per_school_count = pd.read_csv("Resources/schools_complete.csv")

result = per_school_count.groupby("school_name")["size"].sum()

print(result)
```

```python
# Calculate the total school budget and per capita spending per school
per_school_budget = pd.read_csv("Resources/schools_complete.csv")
per_capita_budget = pd.read_csv("Resources/students_complete.csv")

total_budget = per_school_budget.groupby("school_name")["budget"].sum()

total_students = per_capita_budget.groupby("school_name")["Student ID"].count()

capita_budget = total_budget / total_students

print("Total School Budget:")
print(total_budget)
print("------------")
print("Per Capita Spending:")
print(capita_budget)
```

```python
# Calculate the average test scores per school
schools = pd.read_csv("Resources/schools_complete.csv")
students = pd.read_csv("Resources/students_complete.csv")

merged = students.merge(schools, on="school_name", how="inner")

per_school_reading = merged.groupby("school_name")["reading_score"].mean()

per_school_math = merged.groupby("school_name")["math_score"].mean()

print("Average Reading Scores:")
print(per_school_reading)
print("-----------")
print("Average Math Scores:")
print(per_school_math)
```

```python
# Calculate the number of students per school with math scores of 70 or higher
school_avg = pd.read_csv("Resources/schools_complete.csv")
student_avg = pd.read_csv("Resources/students_complete.csv")

merged_df = student_avg.merge(school_avg, on="school_name", how="inner")

math_scores = merged_df[merged_df["math_score"] >= 70]

math_avg = math_scores.groupby("school_name")["Student ID"].count()

print("Math Scores >= 70:")
print(math_avg)
```

```python
# Calculate the number of students per school with reading scores of 70 or higher
school_avgs = pd.read_csv("Resources/schools_complete.csv")
student_avgs = pd.read_csv("Resources/students_complete.csv")

merges = student_avgs.merge(school_avg, on="school_name", how="inner")

reading_scores = merges[merges["reading_score"] >= 70]

reading_avg = reading_scores.groupby("school_name")["Student ID"].count()

print("Reading Scores >= 70:")
print(reading_avg)
```

```python
# Use the provided code to calculate the number of students per school that passed both math and reading with scores of 70 or higher
school_avgs = pd.read_csv("Resources/schools_complete.csv")
student_avgs = pd.read_csv("Resources/students_complete.csv")

merges = student_avgs.merge(school_avg, on="school_name", how="inner")

both_scores = merges[(merges["reading_score"] >= 70) & (merges["math_score"] >= 70)]

both_passing = both_scores.groupby("school_name")["Student ID"].count()

print("Math & Reading >= 70:")
print(both_passing)
```

```python
# Use the provided code to calculate the passing rates
school_avgs = pd.read_csv("Resources/schools_complete.csv")
student_avgs = pd.read_csv("Resources/students_complete.csv")

merges = student_avgs.merge(school_avg, on="school_name", how="inner")

both_scores = merges[(merges["reading_score"] >= 70) & (merges["math_score"] >= 70)]
both_pass = both_scores.groupby("school_name")["Student ID"].count()

students_per_school = merges.groupby("school_name")["Student ID"].count()

passing_rates = (both_pass / students_per_school) * 100

print("Passing Rates:")
print(passing_rates)
```

```python
# Create a DataFrame called `per_school_summary` with columns for the calculations above.
per_school_summary = pd.DataFrame({
    "School Name": school_avgs["school_name"],
    "Total Students": school_avgs["size"],
    "School Type": school_avgs["type"],
    "Total Budgets": total_budget,
    "Per Capita Budgets": capita_budget,
    "Average Math Score": per_school_math,
    "Average Reading Score": per_school_reading,
    "Average Math Score >= 70": math_avg,
    "Average Reading Score >= 70": reading_avg,
    "% Overall Passing": both_passing
})

print(per_school_summary)
```

## Scores by School Spending

```python
# Establish the bins 
spending_bins = [0, 585, 630, 645, 680]
labels = ["<$585", "$585-630", "$630-645", "$645-680"]
```

```python
# Create a copy of the school summary since it has the "Per Student Budget" 
school_spending_df = per_school_summary.copy()
```

```python
# Use `pd.cut` to categorize spending based on the bins.
school_spending_df["Spending Ranges (Per Student)"] = pd.cut(per_school_summary["Per Capita Budgets"], bins=spending_bins, labels=labels)
print(school_spending_df)
```

```python
#  Calculate averages for the desired columns. 
spending_math_scores = school_spending_df.groupby(["Spending Ranges (Per Student)"])["Average Math Score"].mean()
spending_reading_scores = school_spending_df.groupby(["Spending Ranges (Per Student)"])["Average Reading Score"].mean()
spending_passing_math = school_spending_df.groupby(["Spending Ranges (Per Student)"])["Average Math Score >= 70"].mean()
spending_passing_reading = school_spending_df.groupby(["Spending Ranges (Per Student)"])["Average Reading Score >= 70"].mean()
overall_passing_spending = school_spending_df.groupby(["Spending Ranges (Per Student)"])["% Overall Passing"].mean()

per_school_summary["% Overall Passing"] = (per_school_summary["Average Math Score >= 70"] + per_school_summary["Average Reading Score >= 70"]) / 2
overall_spending = school_spending_df.groupby(["Spending Ranges (Per Student)"])["% Overall Passing"].mean()
```

```python
# Assemble into DataFrame
spending_summary = pd.DataFrame({
    "Average Math Score": spending_math_scores,
    "Average Reading Score": spending_reading_scores,
    "% Passing Math": spending_passing_math,
    "% Passing Reading": spending_passing_reading,
    "% Overall Passing": overall_passing_spending
})

# Display results
print(spending_summary)
```

## Scores by School Size

```python
# Establish the bins.
size_bins = [0, 1000, 2000, 5000]
labels = ["Small (<1000)", "Medium (1000-2000)", "Large (2000-5000)"]
```

```python
# Categorize the spending based on the bins
# Use `pd.cut` on the "Total Students" column of the `per_school_summary` DataFrame.

per_school_summary["School Size"] = pd.cut(per_school_summary["Total Students"], bins=size_bins, labels=labels)

print(per_school_summary)
```

```python
# Calculate averages for the desired columns. 
size_math_scores = per_school_summary.groupby(["School Size"])["Average Math Score"].mean()
size_reading_scores = per_school_summary.groupby(["School Size"])["Average Reading Score"].mean()
size_passing_math = per_school_summary.groupby(["School Size"])["Average Math Score >= 70"].mean()
size_passing_reading = per_school_summary.groupby(["School Size"])["Average Reading Score >= 70"].mean()
size_overall_passing = per_school_summary.groupby(["School Size"])["% Overall Passing"].mean()
```

```python
# Create a DataFrame called `size_summary` that breaks down school performance based on school size (small, medium, or large).
# Use the scores above to create a new DataFrame called `size_summary`

size_summary = pd.DataFrame({
    "Average Math Score": size_math_scores,
    "Average Reading Score": size_reading_scores,
    "Average Math Score >= 70": size_passing_math,
    "Average Reading Score >= 70": size_passing_reading,
    "% Overall Passing": size_overall_passing
})

# Display results
print(size_summary)
```
